# Question 1 part A

```python
import timeit

start = timeit.default_timer()

#Time how long fdist runs
fdist = nltk.FreqDist(char for char in text if char.isalpha())
stop = timeit.default_timer()

print (stop - start) 
```

```python
#Running time of fdist = 0.31 s
```

```python
unigrams = list(fdist.keys())
len(unigrams)
```

```python

fdist
```

```python
fdist_len = sum(fdist.values())
fdist_len 
```

```python
prob_ch = {} 
for k,v in fdist.items():
    prob_ch[k] = v/fdist_len
```

```python
prob_ch
```

```python
import math
import timeit

start = timeit.default_timer()

#Time how long entropy compute takes

entropy = 0
for ch in prob_ch.values():
    entropy -= ch * math.log(ch, 2)

stop = timeit.default_timer()

print (stop - start) 
```
